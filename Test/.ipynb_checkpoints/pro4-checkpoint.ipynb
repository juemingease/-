```python
import pandas as pd
import numpy as np
# 绘图
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
df = pd.read_excel(r'.\Sheet12.xlsx').fillna(0)
df = df[["二氧化硅(SiO2)","氧化钠(Na2O)","氧化钾(K2O)","氧化钙(CaO)","氧化镁(MgO)","氧化铝(Al2O3)",\
                "氧化铁(Fe2O3)","氧化铜(CuO)","氧化铅(PbO)","氧化钡(BaO)","五氧化二磷(P2O5)","氧化锶(SrO)",\
               "氧化锡(SnO2)","二氧化硫(SO2)","类型"]]
df1 = df[df["类型"] == 1]
df2 = df[df["类型"] == 2]
df = df.drop(columns='类型')
df1 = df1[["二氧化硅(SiO2)","氧化钠(Na2O)","氧化钾(K2O)","氧化钙(CaO)","氧化镁(MgO)","氧化铝(Al2O3)",\
                "氧化铁(Fe2O3)","氧化铜(CuO)","氧化铅(PbO)","氧化钡(BaO)","五氧化二磷(P2O5)","氧化锶(SrO)",\
               "氧化锡(SnO2)","二氧化硫(SO2)"]].values
df2 = df2[["二氧化硅(SiO2)","氧化钠(Na2O)","氧化钾(K2O)","氧化钙(CaO)","氧化镁(MgO)","氧化铝(Al2O3)",\
                "氧化铁(Fe2O3)","氧化铜(CuO)","氧化铅(PbO)","氧化钡(BaO)","五氧化二磷(P2O5)","氧化锶(SrO)",\
               "氧化锡(SnO2)","二氧化硫(SO2)"]].values
df1 = df1.T
df2 = df2.T
```

```python
def normalize(data):
    return (data-data.min())/(data.max()-data.min())
B = normalize(df1)
B
```

```python
b = B

# 计算灰色关联系数
def grey_relation_coefficient(b, b0, rho):
    n,m = b.shape
    min1 = np.min(np.abs(b0 - b))
    max1 = np.max(np.abs(b0 - b))
    G = np.zeros((n, m))
    for s in range(n):
        for k in range(m):
            d = np.abs(b0[k] - b[s, k])
            G[s, k] = (min1 + rho * max1) / (d + rho * max1)
    return G

rho = 0.5

# 计算灰色关联度
def grey_relation_degree(G, w=None):
    m, n = G.shape
    if w is None:
        w = np.ones(n) / n
    r = np.dot(G, w)
    return r
num_columns = b.shape[1]
w = np.ones(num_columns) / num_columns

correlation = np.zeros((14,14))
for i in range(B.shape[0]):
    b0 = B[i]
    print(b0)
    
    # 计算灰色关联系数
    G = grey_relation_coefficient(B, b0, rho)
    
    # 计算灰色关联度
    r = grey_relation_degree(G, w)
    print(r,type(r))
    correlation[i]=r
    print(correlation)
    

plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap="Blues", fmt=".2f",xticklabels=df.columns, yticklabels=df.columns)
plt.title("高钾类型文物不同化学成分之间的灰色关联热力图")
plt.show()
```

```python
B = normalize(df2)
b = B
num_columns = b.shape[1]
w = np.ones(num_columns) / num_columns
correlation = np.zeros((14,14))
for i in range(B.shape[0]):
    b0 = B[i]
    print(b0)
    
    # 计算灰色关联系数
    G = grey_relation_coefficient(B, b0, rho)
    
    # 计算灰色关联度
    r = grey_relation_degree(G, w)
    print(r,type(r))
    correlation[i]=r
    print(correlation)
    

plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap="Blues", fmt=".2f",xticklabels=df.columns, yticklabels=df.columns)
plt.title("铅钡类型文物不同化学成分之间的灰色关联热力图")
plt.show()
```
