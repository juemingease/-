# 插值
## 应用
已知函数在某区间内若干点处的值，求函数在该区间内其他点处的值
## 方法
1. 拉格朗日插值\

2. 牛顿插值
## 高次插值的Runge现象
插值多项式的次数超过7时，插值多项式会出现严重的振荡现象。\
避免Runge现象的方法：将插值区间分为若干小区间，在小区间内用低次插值（2次，3次），即分段低次插值，如样条函数插值
## 代码示例


```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt

# 3阶样条插值
x = np.arange(0,26,2)
y = np.array([12,9,9,10,18,24,28,27,25,20,15,15,13])
f1 = interp1d(x,y,kind='cubic')   
#"nearest","zero"为阶梯插值
#"slinear" 线性插值
#"quadratic","cubic" 为2阶、3阶B样条曲线插值
y1 = f1(13)
print(y1)
xnew = np.arange(0,24+1/3600,1/3600)
ynew = f1(xnew)
# 画图
plt.scatter(x,y,marker="+") #原始数据
plt.scatter(xnew,ynew,s=1/3600,c="g",marker="*") # 插值数据
plt.legend(['data','cubic'])
plt.tight_layout()
plt.show()

#拉格朗日插值
f2=lagrange(x,y)
ynew1=f2(xnew)
plt.scatter(x,y,marker="+") #原始数据
plt.scatter(xnew,ynew1,s=1/3600,c="g",marker="*") # 拉格朗日插值
plt.legend(['data','Lagrange'])
plt.tight_layout()
plt.show()
```

```python
# 二维插值
import numpy as np
from matplotlib.pyplot import MultipleLocator
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d

x = np.arange(1, 6, 1) 
y = np.arange(1, 4, 1)
z = np.array(
    [
        [82, 81, 80, 82, 84],
        [79, 63, 61, 65, 81],
        [84, 84, 82, 85, 86]
    ]
)
f1 = interp2d(x,y, z, kind='linear')
xnew = np.arange(1, 5.2, 0.2) 
ynew = np.arange(1, 3.2, 0.2)
znew = f1(xnew, ynew)



# 画图
ax = plt.axes(projection='3d')
ax.patch.set_alpha(0)
# 定义坐标轴
X, Y = np.meshgrid(xnew, ynew)
# 作图
ax.plot_surface(X, Y, znew, cmap='rainbow')  
# 设置坐标间隔
ax.set_zlim(60, 90, 5)
ax.zaxis.set_major_locator(MultipleLocator(5))
# 设置视角
ax.view_init(30, 45)
# 设置边界
plt.subplots_adjust(left=0, bottom=0, right=1, top=1)
plt.show()

```

```python
# 二维插值
import numpy as np
from matplotlib.pyplot import MultipleLocator
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d

x =np.mgrid[1200:4400:400]
y =np.mgrid[1200:4000:400]
z =np.array([
    [1130 ,1250 ,1280 ,1230 ,1040 ,900  ,500  ,700 ],
    [1320 ,1450 ,1420 ,1400 ,1300 ,700  ,900  ,850 ],
    [1320 ,1450 ,1420 ,1400 ,1300 ,700  ,900  ,850 ],
    [1500 ,1200 ,1100 ,1550 ,1600 ,1550 ,1380 ,1070],
    [1500 ,1200 ,1100 ,1550 ,1600 ,1550 ,1380 ,1070],
    [1500 ,1550 ,1600 ,1550 ,1600 ,1600 ,1600 ,1550],
    [1480 ,1500 ,1550 ,1510 ,1430 ,1300 ,1200 ,980 ]
])
f1 = interp2d(x,y,z, kind='cubic')
xnew =np.mgrid[1200:4010:10]
ynew =np.mgrid[1200:3610:10]
znew=f1(xnew,ynew)
# 画图
ax=plt.subplot(111,projection='3d')
# 定义坐标轴
X, Y = np.meshgrid(x, y)
XN, YN = np.meshgrid(xnew, ynew)
# 作图
ax.plot_surface(XN, YN, znew, cmap='rainbow')  
# 设置坐标间隔
ax.set_zlim(400,1600,200)
# 设置视角
ax.view_init(30, 45)
# 设置边界
plt.subplots_adjust(left=0, bottom=0, right=1, top=1)
# 加入原散点
ax.scatter(X,Y,z,c='r',marker='^')
plt.show()


C=plt.contour(xnew,ynew,znew,10,cmap='viridis')
plt.clabel(C, inline=True, fontsize=10, fmt="%d")
plt.colorbar(C)
```

```python
# 散乱点插值
from scipy.interpolate import griddata
import numpy as np

x = np.array([129,140,103.5,88,185.5,195,105.5,157.5,107.5,77,81,162,162,117.5])
y = np.array([7.5,141.5,23,147,22.5,137.5,85.5,-6.5,-81,3,56.5,-66.5,84,-33.5])
z = np.array([-4,-8,-6,-8,-6,-8,-8,-9,-9,-8,-8,-9,-4,-9])
grid_x, grid_y = np.mgrid[75:200.5:0.5, -70:150.5:0.5]

grid_z0 = griddata((x,y), z, (grid_x, grid_y), method='nearest')
grid_z1 = griddata((x,y), z, (grid_x, grid_y), method='linear')
grid_z2 = griddata((x,y), z, (grid_x, grid_y), method='cubic')

plt.figure()
ax1 = plt.subplot2grid((2,2), (0,0), projection='3d')
ax1.plot_surface(grid_x, grid_y, grid_z0, color = "c")
ax1.set_xlim3d(75,200)
ax1.set_ylim3d(-70,150)
ax1.set_zlim3d(-10,0)
ax1.set_title('nearest')

ax2 = plt.subplot2grid((2,2), (0,1), projection='3d')
ax2.plot_surface(grid_x, grid_y, grid_z1, color = "c")
ax1.set_xlim3d(75,200)
ax1.set_ylim3d(-70,150)
ax1.set_zlim3d(-10,0)
ax2.set_title('linear')

ax3 = plt.subplot2grid((2,2), (1,0), projection='3d')
ax3.plot_surface(grid_x, grid_y, grid_z2,cmap=plt.get_cmap('rainbow'))
ax1.set_xlim3d(75,200)
ax1.set_ylim3d(-70,150)
ax1.set_zlim3d(-10,0)
ax3.set_title('cubic')

ax4 = plt.subplot2grid((2,2), (1,1), projection='3d')
ax4.scatter(x, y,z,  c= "b")
ax1.set_xlim3d(75,200)
ax1.set_ylim3d(-70,150)
ax1.set_zlim3d(-10,0)
ax4.set_title('org_points')

plt.tight_layout()
plt.show()

C=plt.contour(grid_x, grid_y, grid_z2,[-5])
```
