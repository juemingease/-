# 聚类分析
## 定义
将一个数据集通过相似度划分为若干组或类，相似度是基于数据描述属性的取值来确定的，通常用**距离**来表示。\
用于对样本结构做初步的评价
## 应用
1. 市场分析
2. 万维网
3. 图像处理
4. 模式识别
5. 孤立点检测
## 分类
1. Q型：对样品分类（常用）
2. R型：对变量（指标）分类
## 距离计算
1. 欧式距离（常用）$$d(x_i,x_j)=[\sum \limits _{k=1}^p(x_{ik}-x_{jk})^2]^{1/2}$$
2. 绝对距离（少用）
3. 明氏距离
4. 切氏距离
5. 方差加权距离
6. 马氏距离（原数据标准化以后的欧式距离）
7. 兰氏距离
8. 杰氏距离
## 变量间的相似度量——相似系数
取值范围（-1，1）负相关性->不相关->正相关性
1. 夹角余弦
2. 相关系数
## 类间距离
1. 最短距离
2. 最长距离
3. 类平均距离（距离的平均值）
4. 重心距离（平均值的距离）
5. 离差平均和的距离（少用）

## 例题



```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import AgglomerativeClustering

data = np.array(
    [
        [7.90,39.77,8.49,12.94,19.27,11.05,2.04,13.29],
        [7.68,50.37,11.35,13.30,19.25,14.59,2.75,14.87],
        [9.42,27.93,8.20,8.14,16.17,9.42,1.55,9.76],
        [9.16,27.98,9.01,9.32,15.99,9.10,1.82,11.35],
        [10.06,28.64,10.52,10.05,16.18,8.39,1.96,10.81]
    ]
)

z=linkage(data,"single" ,metric='euclidean')
# 类间距离为最短距离，距离计算使用欧式距离
print(z) # 聚类过程

 #画聚类图
fig= plt.subplots(figsize=(10,9)) # 图片尺寸
dendrogram(z)
plt.title("Hierachial Clustering Dendrogram")
plt.xlabel("Cluster label")
plt.ylabel("Distance")
plt.show()

# 如果仅有二维，可以画散点图如下
'''
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
#linkage模式可以调整,n_cluser可以调整
print(cluster.fit_predict(data))
plt.figure(figsize=(10, 7))  
plt.scatter(data[0],data[1], c=cluster.labels_)
plt.show()
'''
```
