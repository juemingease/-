# 拟合
## 与插值区别
1. 插值过已知点，拟合不一定
2. 插值求函数值，拟合求函数关系
## 计算
1. 线型的选择
2. 线型中的参数计算
线性：最小二乘\
非线性：高斯-牛顿迭代
## 拟合方法
1. 多项式拟合
2. 非线性拟合

```python
import numpy as np
from numpy import polyfit,poly1d
import matplotlib.pyplot as plt

x = np.arange(1,13,1)
y = np.array([5,8,9,15,25,29,31,30,22,25,27,24])
coeff = polyfit(x,y,9) # 拟合参数
f = poly1d(coeff) # 将参数导入多项式
xp = np.arange(1,12.1,0.1)
yp = f(xp)
plt.scatter(x,y,c='r',marker='*',label='data')
plt.plot(xp,yp,c='b',label='fit')
plt.legend(loc='upper right')
plt.title('Data Fitting')
plt.show()
```

```python
# 非线性拟合
# 解决常见的非线性拟合可以使用matlab的cftools
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from math import exp

x = np.arange(1,17,1)
y = np.array([4,6.4,8,8.8,9.22,9.5,9.7,9.86,10.00,10.2,10.32,10.42,10.5,10.55,10.58,10.6])
# 待拟合函数
def model(x, a, b, c, d, e ):
    return a * np.exp(-x / b) + c * np.exp(-x / d) + e
popt, pcov = curve_fit(model, x, y)
print(popt)
xnew = np.arange(1,16.1,0.1)
ynew = model(xnew,*popt)
plt.scatter(x,y,c='r',marker='*')
plt.plot(xnew,ynew)
plt.show()
```
